```python
import pandas as pd
import numpy as np
data = pd.read_csv("train.csv", index_col="Loan_ID")
```

```python
data
```

```python
data.loc[(data["Gender"] == "Female") & (data["Education"] == "Not Graduate") & (data["Loan_Status"] == "Y"), ["Gender", "Education", "Loan_Status"]]
```

```python
def num_missing(x):
    return sum(x.isnull())
```

```python
data.apply(num_missing)
```

```python
data.apply(num_missing, axis=1).head()
```

```python
from scipy.stats import mode
mode(data['Gender'].astype(float))
```

```python
mode(data['Gender'].astype(str)).mode[0]
```

```python
data["Gender"].fillna(mode(data['Gender'].astype(str)).mode[0], inplace = True)
```

```python
data['Married'].fillna(mode(data['Married'].astype(str)).mode[0], inplace=True)
data['Self_Employed'].fillna(mode(data['Self_Employed'].astype(str)).mode[0], inplace=True)
```

```python
data.apply(num_missing)
```

```python
impute_grp = data.pivot_table(values = ['LoanAmount'], index = ["Gender","Married","Self_Employed"], aggfunc=np.mean)
```

```python
print(impute_grp)
```

```python
for i, row in data.loc[(data["LoanAmount"].isnull()),:].iterrows():
    ind = tuple([row["Gender"], row["Married"], row["Self_Employed"]])
    data.loc[i, "LoanAmount"] = impute_grp.loc[ind].values[0]
    
```

```python
data.apply(num_missing)
```

```python
pd.crosstab(data["Credit_History"],data["Loan_Status"], margins=True).apply(perce)
```
